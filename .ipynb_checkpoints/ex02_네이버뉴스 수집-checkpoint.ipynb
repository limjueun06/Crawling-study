### 데이터 수집 순서
1. 뉴스페이지의 url 주소 수집 및 리스트 저장
2. 리스트와 반복문을 이용해서 각 뉴스 페이지의 문서를 요청
3. 각 뉴스 페이지 문서 변환
4. 변환된 문서 내 뉴스 제목과 내용 수집

```python
# 뉴스페이지 url 
news_section_url = 'https://news.naver.com/section/105'

result = req.get(news_section_url)

html = bs(result.text, 'lxml')
```

```python
news_list = html.select('a.sa_text_title')
news_list

# select -> 리스트 반환 -> 인덱스 번호 사용가능
```

```python
news_list[0].text
```

```python
news_list[0].get('href')
```

```python
# 실습) 방법1
# 현재 news_list에 있는 모든 경로를 리스트로 저장해보기
# 경로는 href 속성 / 반복문 사용
href_list = []

for h in news_list :
    href_list.append(h.get('href'))

href_list
```

```python
len(news_list)
```

```python
# 실습) 방법2
# 현재 news_list에 있는 모든 경로를 리스트로 저장해보기
# 경로는 href 속성 / 반복문 사용
href_list = []

for h in range(0, 44) :
    href_list.append(news_list[h].get('href'))

href_list
```

```python
for h in range(len(news_list)) :
    href_list.append(news_list[h].get('href'))

href_list

[h.get('href') for h in news_list]
```

```python
# 리스트 컴프리핸션
# news.get('href') | for news in news_list : for문이 먼저, 다음 실행문 실행
href_list = [news.get('href') for news in news_list]
href_list
```

```python
# 위에서 살펴본 뉴스제목, 본문, 경로 3가지를 한번에 가져와 정리.

data_list = [] # 뉴스제목, 본문, 경로를 저장하는 리스트
for href in href_list : # 경로데이터로 반복문 실행
    result = req.get(href) # 경로를 집어 넣어서 요청하겠다.

    html = bs(result.text, 'lxml') # 문자열 -> HTML로 파싱(변환)

    title = html.select_one('#title_area > span').text # 제목
    content = html.select_one('#dic_area').text.replace('\n\n\n', '').strip() # 뉴스 본문

    data_list.append([title, content, href])
data_list    
```

```python
import pandas as pd
news_df = pd.DataFrame(data_list, columns = ['뉴스제목', '내용', '페이지주소'])
news_df
```

```python
# to_csv('./경로/파일이름.csv', encoding = 'euc-kr' / 'utf-8', index = True / False)
# 데이터프레임을 csv라는 형식의 파일로 저장.
# encoding : 데이터 저장 방식
# index : True = 기존 인덱스 저장 / False = 기존 인덱스 저장 X
news_df.to_csv('IT_과학 뉴스데이터 수집.csv', index = False)
```
