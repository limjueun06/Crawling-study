```python
def check_age(age):
    if age < 0:
        # ValueError라는 에러 타입과 함께 메시지를 전달합니다.
        raise ValueError("나이는 0보다 작을 수 없습니다! 입력된 값: " + str(age))
    else:
        print(f"입력된 나이는 {age}세입니다.")

# 정상 실행
check_age(25)

# 에러 발생 시뮬레이션
check_age(-5)
```

```python
def withdraw_money(balance, amount):
    try:
        print(f"--- 출금 시도: {amount}원 (현재 잔액: {balance}원) ---")
        
        if amount > balance:
            # 잔액 부족 시 강제로 에러 발생
            raise Exception(f"잔액이 부족합니다. (부족 금액: {amount - balance}원)")
        
        if amount <= 0:
            # 유효하지 않은 금액 입력 시 에러 발생
            raise ValueError("출금 금액은 0원보다 커야 합니다.")

    except ValueError as e:
        # 특정 에러(ValueError)를 개별적으로 처리
        print(f"⚠️ 입력 오류: {e}")

    except Exception as e:
        # raise로 던진 Exception이나 기타 모든 에러를 처리
        # 'as e'를 통해 에러 메시지를 변수로 받아 출력할 수 있습니다.
        print(f"❌ 시스템 오류 발생: {e}")

    else:
        # 에러가 발생하지 않았을 때만 실행되는 구간
        balance -= amount
        print(f"✅ 출금 성공! 남은 잔액은 {balance}원입니다.")

    finally:
        # 에러 발생 여부와 상관없이 무조건 마지막에 실행
        print("📢 거래를 종료합니다. 이용해 주셔서 감사합니다.\n")

# 1. 정상 거래 케이스
withdraw_money(10000, 5000)

# 2. 잔액 부족 케이스 (Exception 발생)
withdraw_money(10000, 15000)

# 3. 잘못된 값 입력 케이스 (ValueError 발생)
withdraw_money(10000, -100)
```
