```python
# 필요한 라이브러리 가져오기
!pip install selenium
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import random
import pandas as pd

# undetected-chromedriver : 자동화탐지우회, 크롤링차단방지
# undetected-chromedriver 설치
!pip install undetected-chromedriver
!pip show undetected-chromedriver

# 크롬 드라이브 불러오기
import undetected_chromedriver as uc

# 크롬 드라이브 열어보기
driver = uc.Chrome()

# 지마켓 베스트메뉴 주소 저장, 실행
url = 'https://www.gmarket.co.kr/'
driver.get(url)

# 지마켓 검색창에 "국내산 찰보리 경주빵" 검색
search = driver.find_element(By.CSS_SELECTOR, '#form__search-keyword')
search.send_keys("국내산 찰보리 경주빵")
search.send_keys(Keys.ENTER)
time.sleep(random.uniform(0.5, 1))

# 상품 수집을 위한 태그 수집
# 상품 갯수확인
item = driver.find_elements(By.CSS_SELECTOR, 'div.box__item-title span.text__item-title a')
len(item)

# 첫번째 상품 클릭
item[0].click()
time.sleep(random.uniform(0.5, 1))

# 새 탭에서 상품상세페이지가 열림 (새탭으로 이동)
# driver.switch_to.window(driver.window_handles[0]) -> 검색결과페이지
driver.switch_to.window(driver.window_handles[1]) # -> 상품상세페이지
time.sleep(random.uniform(0.5, 1))

# 상품평 클릭
review_tab = driver.find_element(By.CSS_SELECTOR, '#container > div.vip-tabwrap.uxetabs > div.vip-tabnavi.uxeposfix > ul > li:nth-child(2) > a')
review_tab.click()
time.sleep(random.uniform(0.5, 1))

# 프리미엄 리뷰(상품평)수집
pre_review_list = []

# 리뷰 총 페이지 계산
pre_page_move_btn = driver.find_elements(By.CSS_SELECTOR, "#premium-pagenation-wrap > div.board_paging > div > ul > li > a")
len(pre_page_move_btn)

# 반복문을 활용해서 모든 페이지를 돌아보고 리뷰를 가져오자
try :
    for i in range(len(pre_page_move_btn)) : # 15번 반복
        # 버튼 선택자 찾기
        pre_btn = driver.find_element(By.CSS_SELECTOR, "#premium-pagenation-wrap > div.board_paging > div > button")
    
        # 버튼 클릭
        pre_btn.click()
        time.sleep(random.uniform(0.5, 1))
    
        # 페이지 이동
        pre_page_move_btn = driver.find_elements(By.CSS_SELECTOR, "#premium-pagenation-wrap > div.board_paging > div > ul > li > a")
        pre_page_move_btn[i].click()
        time.sleep(random.uniform(0.5, 1))
    
        # 리뷰 수집
        pre_review = driver.find_elements(By.CSS_SELECTOR, "#premium-wrapper > table > tbody > tr > td.comment-content > a > p.con")
    
        for pre_rv in pre_review :
            print(pre_rv.text) # 확인용 리뷰 출력
            pre_review_list.append(pre_rv.text)
        time.sleep(random.uniform(0.5, 1))
except :
    print('에러로 종료!') # except 다음에 코드를 넣어줘야함.
print('수집종료')

review_dict = {'리뷰 수집' : pre_review_list}
df = pd.DataFrame(review_dict)

df.to_csv('프리미엄 리뷰수집.csv', index = False)

driver.quit()
```
