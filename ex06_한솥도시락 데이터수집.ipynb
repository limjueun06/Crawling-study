- 이때 도시락 가격정보를 활용해서 학생들이 만족할만한 도시락을 합리적으로 구매하기위해 데이터를 확보하고자 한다.

```python
# 필요한 라이브러리 불러오기
import pandas as pd
import time
import random
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
# 사이트주소 가져오기
url = 'https://www.hsd.co.kr/menu/menu_list'
```

```python
# 브라우저 실행
# 컴퓨터야! 크롬 드라이버 켜줘
driver = wb.Chrome()
```

```python
# 한솥 메뉴 페이지로 이동하기
driver.get(url)
```

```python
# 더보기 버튼 눌러서 모든 도시락 메뉴 보이도록 만들기

while True :
    try : 
        more_btn = driver.find_element(By.CSS_SELECTOR, 'a.c_05')
        more_btn.click()
        time.sleep(1)
    except :
        print('더보기 버튼 끝!')
        break
```

```python
### 실습)
# 도시락 이름 / 가격 가져와서 각각 리스트에 저장해보자.

# 빈 리스트 만들기
# 반복문 구현해서 이름과 가격을 각각 리스트에 집어넣기
```

```python
name_list = []

name = driver.find_elements(By.CSS_SELECTOR, '.h.fz_03')
for i in name :
    name_list.append(i.text)

print(name_list)
len(name_list)
```

```python
price_list = []

price = driver.find_elements(By.CSS_SELECTOR, '.item-price')
for j in price :
    price_list.append(j.text)

print(price_list)
len(price_list)
```

```python
# 빈 리스트 만들기
name_list = []
price_list = []

# 선택자로 이름과 가격을 가져오기
name = driver.find_elements(By.CSS_SELECTOR, "div.item-text h4.h.fz_03")
price = driver.find_elements(By.CSS_SELECTOR, "div.item-price strong")

# 반복문 구현해서 이름과 가격을 각각 리스트에 집어넣기
print(len(name))
print(len(price))

for i in range(len(name)) :
    name_list.append(name[i].text)
    price_list.append(price[i].text)
```

```python
# 코드 통합 후 크롬 드라이버 종료하고 자료 저장하기
# 1. 필요한 라이브러리 불러오기
import pandas as pd
import time
import random
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
# =================================================
# 2. 사이트주소 가져오기
url = 'https://www.hsd.co.kr/menu/menu_list'

# 3. 브라우저 실행
# 컴퓨터야! 크롬 드라이버 켜줘
driver = wb.Chrome()

# 4. 한솥 메뉴 페이지로 이동하기
driver.get(url)
time.sleep(1)

# 5. 더보기 버튼 눌러서 모든 도시락 메뉴 보이도록 만들기

while True :
    try : 
        more_btn = driver.find_element(By.CSS_SELECTOR, 'a.c_05')
        more_btn.click()
        time.sleep(1)
    except :
        print('더보기 버튼 끝!')
        break

# 6. 빈 리스트 만들기
name_list = []
price_list = []

# 7. 선택자로 이름과 가격을 가져오기
name = driver.find_elements(By.CSS_SELECTOR, "div.item-text h4.h.fz_03")
price = driver.find_elements(By.CSS_SELECTOR, "div.item-price strong")

# 8. 반복문 구현해서 이름과 가격을 각각 리스트에 집어넣기
print(len(name))
print(len(price))

for i in range(len(name)) :
    name_list.append(name[i].text)
    price_list.append(price[i].text)

# 9. 만들어진 리스트로 딕셔너리 구현 후 데이터프레임 만들기
menu_dict = {'메뉴 이름' : name_list, '가격 정보' : price_list}
df = pd.DataFrame(menu_dict)

# 10. 크롬 드라이버 종료
driver.quit()

# 11. 데이터프레임 저장하기
df.to_csv('한솥도시락목록.csv', index = False)
```
