### req라이브러리를 이용해서 여러 페이지에서 데이터를 수집해보자.

```python
# 필요한 라이브러리 가져오기
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests as req
```

- 선택자를 분석하는 방법
- 페이지를 이동하는 방법

```python
url = 'https://search.zdnet.co.kr/'

result = req.get(url,
                 params = {'kwd' : 'AI'})

print(result)
```

### 주목할 부분
- reqests로 진행하는 요청은 해당 URL의 내용만 받아온다.
- page가 변경되는 경우 URL도 변경된다.
- URL의 변화에 포커싱을 맞추자 

### 사람이 사이트에서 자료를 수집 할때 어떻게 하는지 봅시다.
- 두번째 페이지를 수집하려면?
- 사람이라면 두번째 페이지를 클릭
- 2번 페이지 주세요! 요청 > URL을 통해서 새로운 사이트를 요청
- URL을 보면 &pageno=2 로 변경된것을 살펴 볼 수 있다.
- 요청을 하고 수집이 끝나면 페이지 번호를 바꿔서 보내자.
- 요청 -> 수집 -> 요청 -> 수집 ... = 반복되는 내용 (반복문을 활용하자)

### 반복문을 작성하는 방법
- 반복문은 일정한 코드가 반복하면서 동작하는 문법
- 크롤링에서는 정해진 범위 / 횟수 만큼 반복되는 일이 많음 > for문 많이 사용.
- 반복문을 작성할때 무슨 작업을 먼저 해야하고 반복이 어디서 일어나는지 생각.

```python
# 1. 타이틀이 저장되는 리스트와 URL이 저장되는 리스트를 만들어주자
title_list = []
url_list = []

# 2. 반복문 설정
for i in range(1, 6) : # 1 ~ 5라는 숫자 생성
    result = req.get(f'https://search.zdnet.co.kr/?kwd=ai&area=&pageno={i}&term=') # url 페이지 요청(페이지변경)
    soup = bs(result.text, 'lxml') # 데이터 파싱
    # 3. 타이틀과 URL을 가져오는 방법을 설정하자
    # 뉴스 헤드라인 가져오기
    title = soup.select('body > div.contentWrapper > div > div.left_cont > section > div > div.assetText > a > h3')
    # url 가져오기
    news_url = soup.select('body > div.contentWrapper > div > div.left_cont > section > div > div.assetText > a')

    # 가져온 데이터를 리스트에 채워주기
    for j in range(len(title)) :
        # 타이틀 리스트에 값을 채워주자
        title_list.append(title[j].text)
        url_list.append('https://zdnet.co.kr' + news_url[j]['href'])
```

```python
# 데이터 갯수 매칭 확인하기
len(title_list), len(url_list)
```

```python
# 데이터 프레임으로 만들고 나서, csv 파일로 저장후 확인
data = {'뉴스제목' : title_list,
         '링크' : url_list }
rs = pd.DataFrame(data)
rs.to_csv('뉴스데이터.csv', index = False)
```
