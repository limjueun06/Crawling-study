```python
# 필요한 라이브러리 불러오기

# 브라우저를 제어하는 라이브러리(컴퓨터용 브라우저) -★필수
from selenium import webdriver as wb
# 선택자를 구분지어주는 라이브러리 -★필수
from selenium.webdriver.common.by import By
# 컴퓨터용 키보드 라이브러리 -> enter / Esc등 기능을 입력 할 때 사용 - 선택(자주사용)
from selenium.webdriver.common.keys import Keys
```

```python
# 1. 컴퓨터용 브라우저 켜기
driver = wb.Chrome()
```

```python
# 2. 원하는 사이트 요청하기
driver.get('https://www.naver.com')
```

```python
# 3. 검색창에 "베네수엘라" 검색하기
# 검색을 진행하기 위해 컴퓨터에게 검색창을 알려주기
search = driver.find_element(By.CSS_SELECTOR, '#query')
```

### 요소를 찾는 방법
1. driver.find_element(s)(By.CSS_SELECTOR, '선택자')
2. soup.select(s)와 같은 기능
3. BeautifulSoup을 불러오지 않았다!!
    - BeautifulSoup을 불러온 이유 : str -> html로 파싱하기 위해서
4. driver는 브라우저 그 자체 -> 가지고 있는 데이터(이미 HTML)
5. selenium에서는 모든 정보를 가지고 있는 주체가 driver임을 기억하자!

### 다시 검색을 하려니 오류가 발생
- 브라우저는 페이지가 바뀌면 모든 내용이 초기화 된다.
- 페이지가 바뀌면 태그를 다시 찾아서 지정해야한다.
- tip! 통신 방식에 따라 다르다.
- 동기 통신 : 전체 페이지가 바뀐다 -> 전체 데이터가 초기화된다.
- 비동기 통신 : 전체 페이지에서 특정 부분만 바뀐다 -> 특정부분만 초기화

```python
# 검색창 재검색 후 검색을 진행 -> click을 활용
search = driver.find_element(By.CSS_SELECTOR, '#query')
search.send_keys('동계올림픽')
# 버튼 클릭하기
btn = driver.find_element(By.CSS_SELECTOR, '.btn_search')
btn.click()
```

```python
# 브라우저 정리하기
# 프로세스를 정리해야 메모리를 효율적으로 쓸 수 있다.
driver.quit()
```

### 코드 통합을 통해 최종 완성본 만들기 -> 오후 진행

```python
# 필요한 라이브러리 불러오기

# 브라우저를 제어하는 라이브러리(컴퓨터용 브라우저) -★필수
from selenium import webdriver as wb
# 선택자를 구분지어주는 라이브러리 -★필수
from selenium.webdriver.common.by import By
# 컴퓨터용 키보드 라이브러리 -> enter / Esc등 기능을 입력 할 때 사용 - 선택(자주사용)
from selenium.webdriver.common.keys import Keys
# 총 10번 반복해서 스크롤을 내려주자.
import time # 시간을 제어하는 라이브러리
import random # 난수를 생성하는 라이브러리
import pandas as pd

# ================================================================================

# 1. 컴퓨터용 브라우저 켜기
driver = wb.Chrome()

# 2. 원하는 사이트 요청하기
driver.get('https://www.naver.com')
time.sleep(1) # 추가) 1초 대기


# 3. 검색창에 "베네수엘라" 검색하기
# 검색을 진행하기 위해 컴퓨터에게 검색창을 알려주기
search = driver.find_element(By.CSS_SELECTOR, '#query')

# 4. 검색어 입력하기
search.send_keys("베네수엘라")

# 5. 키보드에 enter키 사용하기
search.send_keys(Keys.ENTER)
time.sleep(1) # 추가) 1초 대기

# 6. 뉴스 탭을 찾아서 가보기
news_tab = driver.find_element(By.CSS_SELECTOR, '#lnb > div.lnb_group > div > div.lnb_nav_area._nav_area_root > div > div.api_flicking_wrap._conveyer_root > div:nth-child(1) > a')
news_tab.click()
time.sleep(1) # 추가) 1초 대기

# 7. body 태그 가져와서 스크롤 내리기
body = driver.find_element(By.CSS_SELECTOR, 'body')
body.send_keys(Keys.END)
for i in range(10) : # 0 ~ 9번 반복
    time.sleep(random.uniform(0.5, 1.0)) # sleep() : 0.5에서 1초 사이에 랜덤으로 코드 일시정지
    body.send_keys(Keys.END) # 스크롤 내리기

# 8. 뉴스 타이틀 가져와보기(수정 들어가기)
# 뉴스 타이틀 저장하는 기능으로 수정
title = driver.find_elements(By.CSS_SELECTOR, "a[data-heatmap-target='.tit']")

title_list = [] # 수정) 빈 리스트 추가
for i in title :
    title_list.append(i.text) # 수정) 데이터 리스트 적재

# 9. 브라우저 정리하기
# 프로세스를 정리해야 메모리를 효율적으로 쓸 수 있다.
driver.quit()

# 10. 수집한 데이터 저장하기
df = pd.DataFrame(title_list, columns = ['뉴스 헤드라인'])
df.to_csv('뉴스헤드라인.csv', index = False)
```
